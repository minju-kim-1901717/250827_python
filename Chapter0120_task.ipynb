### 리팩토링 과제

Day4에서 만든 "붕어빵 자동 판매 시스템"은 절차적 코드로 구현되어있습니다.
<br>

Day5에서 배운 함수, 클래스 개념을 적용시켜 객체지향적으로 리팩토링을 진행해주세요


```python
# 붕어빵 가게 클래스
class BungeoppangShop:
    def __init__(self):
        # 재고 초기화
        self.stock = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}
        # 가격 초기화
        self.prices = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}
        # 판매량 초기화
        self.sales = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}

    # 현재 재고 확인
    def check_stock(self):
        for bread, count in self.stock.items():
            print(f"{bread}: {count}개")

    # 주문 처리 메소드
    def process_order(self, bread_type, bread_count):
        # 메뉴에 없는 경우
        if bread_type not in self.stock:
            print("😅 없는 메뉴입니다")
            return

        # 재고 부족한 경우
        elif self.stock[bread_type] < bread_count:
            print("😅 재고가 부족합니다")
            return

        # 주문 가능한 경우
        else:
            self.stock[bread_type] -= bread_count       # 재고 차감
            self.sales[bread_type] += bread_count       # 판매량 증가
            print(f"🥐 {bread_type} {bread_count}개 주문이 완료되었습니다")

    # 관리자 모드 (재고 추가)
    def admin_mode(self):
        print("😈 관리자 모드입니다 😈")
        bread_type = input("추가할 붕어빵 종류: ")
        additional_stock = int(input("추가할 재고량: "))

        # 없는 메뉴인지 확인
        if bread_type not in self.stock:
            print("없는 메뉴입니다")
            return

        # 재고 추가
        self.stock[bread_type] += additional_stock
        print(f"{bread_type} 재고가 {additional_stock}개 추가되었습니다")

    # 총 매출 계산
    def calculate_total_sales(self):
        total_sales = 0
        # 판매량 × 가격 = 매출
        for bread, count in self.sales.items():
            total_sales += count * self.prices[bread]
        print(f"💰 총 매출: {total_sales}원")


# 프로그램 시작 지점
def main():
    # 가게 객체 생성
    shop = BungeoppangShop()
    print("🥐 붕어빵 가게에 오신 것을 환영합니다 🥐")

    # 루프 실행
    while True:
        mode = input("메뉴를 선택하세요 (주문 / 관리자 / 종료): ")

        if mode == "주문":
            print("🥐 어떤 붕어빵을 주문하실건가요")
            print("🫘팥붕어빵 | 🧁슈크림붕어빵 | 🍫초코붕어빵")
            bread_type = input()
            count = int(input("🥐 주문할 개수: "))
            shop.process_order(bread_type, count)   # 주문 처리
            shop.check_stock()                     # 주문 후 재고 확인

        elif mode == "관리자":
            shop.admin_mode()

        elif mode == "종료":
            break

        else:
            print("잘못된 입력입니다.")

    # 프로그램 종료 시 매출 계산
    shop.calculate_total_sales()
    print("🥐 프로그램을 종료합니다. 감사합니다! 🫶")


# main 실행
if __name__ == "__main__":
    main()
```

__BungeoppangShop 클래스__

- __init__(): 초기 재고, 가격, 판매량을 설정합니다.
- check_stock(): 현재 재고를 출력합니다.
- process_order(): 주문을 처리하고 재고를 업데이트합니다.
- admin_mode(): 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales(): 총 매출을 계산하고 출력합니다.
